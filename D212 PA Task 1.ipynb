### PART II: TECHNIQUE JUSTIFICATION

#### B1. Explanation of Clustering Technique<a id='b1'></a>

Clustering works by grouping points from the dataset into different groups, or clusters, by looking for similarities. The user must specify the desired number of clusters that will be used to group the data. An Elbow Graph is typically used to determine the number of recommended cluster groups.

K-means randomly selects K number of points in the data, where K is the number of clusters chosen, representing the center of a cluster.  The center of a cluster is called the centroid.  The distance is measured for each data point to each centroid and assigned to the closest cluster. The next step is to calculate each cluster's mean, which becomes the cluster's new centroid. Repeat the data point assignment to the nearest cluster. Calculating the mean of each cluster and reassigning data points to the closest clusters continues until no point is reassigned.

#### B2. Summary of Technique Assumption<a id='b2'></a>

There are very few assumptions for the k-means clustering model.  The algorithm assumes that the clusters are spherical in shape and that clusters are similar in size (Perceptive Analytics, 2017).

#### D2. Code Execution<a id='d2'></a>

```python
#Importing needed libraries/packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
#Importing medical dataset CSV file
df = pd.read_csv('Files/medical_clean.csv')
```

```python
#Checking dataset shape
print('Data Rows and Columns')
print('-'*21)
print(f'There are {df.shape[0]} rows and {df.shape[1]} columns')
print('-'*100)
print('\n')

#Checking data types and null values
print('Data Info')
print('-'*9)
print(df.info())
print('-'*100)
print('\n')

#Checking data for duplicates
print('Data Duplicate Info')
print('-'*19)
print(df.duplicated().value_counts())
```

```python
#Selecting desired variables to cluster
num_var = ['Age', 'Income', 'Doc_visits']

#Checking statistical information of num_var
df[num_var].describe().T
```

```python
#Creating histograms of each variable in num_var to
plt.figure(1, figsize=(15,6))
n = 0

for x in num_var:
    n+=1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
    sns.histplot(df[x], bins=20, kde=True)
    plt.title('Histogram of {}'.format(x))
plt.show()
```

```python
#Creating Age groups
age1 = df[(df['Age'] >= 18) & (df['Age'] < 30)]
age2 = df[(df['Age'] >= 30) & (df['Age'] < 42)]
age3 = df[(df['Age'] >= 42) & (df['Age'] < 54)]
age4 = df[(df['Age'] >= 54) & (df['Age'] < 66)]
age5 = df[(df['Age'] >= 66) & (df['Age'] < 78)]
age6 = df[(df['Age'] >= 78)]

#Creating x and y values for plot
agex_value = ['18 - 29', '30 - 41', '42 - 53', '54 - 65', '66 - 77', '>=78']
agey_value = [len(age1), len(age2), len(age3), len(age4), len(age5), len(age6)]

#Creating bar plot
plt.figure(1, figsize=(15, 6))
sns.barplot(x =agex_value, y=agey_value)
plt.xlabel('Age Groups')
plt.ylabel('Number of Patients')
plt.title('Number of Patients by Age Group')
plt.show()
```

```python
#Creating Income groups
income1 = df[(df['Income'] >= 0) & (df['Income'] < 25000)]
income2 = df[(df['Income'] >= 25000) & (df['Income'] < 50000)]
income3 = df[(df['Income'] >= 50000) & (df['Income'] < 75000)]
income4 = df[(df['Income'] >= 75000) & (df['Income'] < 100000)]
income5 = df[(df['Income'] >= 100000) & (df['Income'] < 125000)]
income6 = df[(df['Income'] >= 125000) & (df['Income'] < 150000)]
income7 = df[(df['Income'] >= 150000) & (df['Income'] < 175000)]
income8 = df[(df['Income'] >= 175000) & (df['Income'] < 200000)]
income9 = df[(df['Income'] >= 200000)]

#Creating x and y values for plot
incomex_value = ['0 - <25k', '25k - <50k', '50k - <75k', '75k - >100k', '100k - <125k', '125k - <150k', '150k - <175k',
              '175k - <200k', '>=200k']
incomey_value = [len(income1), len(income2), len(income3), len(income4), len(income5), len(income6), len(income7), 
             len(income8), len(income9)]

#Creating bar plot
plt.figure(1, figsize=(15, 6))
sns.barplot(x =incomex_value, y=incomey_value)
plt.xlabel('Income Groups')
plt.ylabel('Number of Patients')
plt.title('Number of Patients by Income Group')
plt.show()
```

```python
#Count plot showing customer count by Doc_visits
plt.figure(1, figsize=(15, 6))
sns.countplot(x = df['Doc_visits'])
plt.xlabel('Doctor Visits Group')
plt.ylabel('Number of Patients')
plt.title('Number of Patients by Doctor Visit Group')
plt.show()
```

```python
#Standardizing 'Age', 'Income' and 'Doc_visit'
X = df[['Age', 'Income', 'Doc_visits']]
standard = df[['Age_S', 'Income_S', 'Doc_visits_S']] = StandardScaler().fit_transform(X)
```

```python
#Creating CSV file for updated df
df.to_csv(r'Task 1/Files/medical_prepared.csv', index=False)
```

```python
#Creating graph to use elbow method for determining number of clusters
inertia = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init='auto')
    kmeans.fit(standard)
    inertia.append(kmeans.inertia_)
plt.figure(figsize=(10,4))
plt.grid()
plt.plot(range(1,10), inertia, linewidth=2, color='green', marker='8')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Cluster Elbow Graph')
plt.show()
```

```python
#Creating k-means analysis with 4 clusters
model = KMeans(n_clusters=4, n_init='auto')
cluster1 = model.fit_predict(standard)
df['Cluster1'] = cluster1
```

```python
#Creater cluster scatter plot of 'Age' and 'Income'
scatter1 = sns.scatterplot(data=df, x='Age', y='Income', hue='Cluster1', palette='tab10')
sns.move_legend(scatter1, 'upper left', bbox_to_anchor=(1,1))
plt.title('4 Cluster Plot')
plt.show()
```

```python
#Calculating silhouette score
silhouette_score(standard, cluster1)
```

```python
#Standardizing 'Age' and 'Income'
X = df[['Age', 'Income']]
standard = df[['Age_S', 'Income_S']]
```

```python
#Creating graph to use elbow method for determining number of clusters
inertia = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init='auto')
    kmeans.fit(standard)
    inertia.append(kmeans.inertia_)
plt.figure(figsize=(10,4))
plt.grid()
plt.plot(range(1,10), inertia, linewidth=2, color='green', marker='8')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Cluster Elbow Graph')
plt.show()
```

```python
#Creating k-means analysis with 3 clusters
model = KMeans(n_clusters=3, n_init='auto')
cluster2 = model.fit_predict(standard)
df['Cluster2'] = cluster2
```

```python
#Creater cluster scatter plot of 'Age' and 'Income'
scatter2 = sns.scatterplot(data=df, x='Age', y='Income', hue='Cluster2', palette='tab10')
sns.move_legend(scatter2, 'upper left', bbox_to_anchor=(1,1))
plt.title('3 Cluster Plot')
plt.show()
```

```python
#Calculating silhouette score
silhouette_score(standard, cluster2)
```

```python
#Creating CSV file for updated df with cluster groups
df.to_csv(r'Task 1/Files/medical_clustered.csv', index=False)
```

```python
#Creating visualizations

#Reviewing readmission by cluster group count
plot = sns.countplot(data=df, x='Cluster2', hue='ReAdmis')
plt.title('Readmission by Patient Count and Cluster Group')
plt.xlabel('Cluster Group')
plt.ylabel('Number of Patients')
sns.move_legend(plot, 'upper left', bbox_to_anchor=(1,1))
plt.show()

#Creating box plot for Age and Cluster2 and ReAdmis
plot1 = sns.boxplot(data=df, x='Cluster2', y='Age', hue='ReAdmis')
plt.title('Readmission by Patient Age and Cluster Group')
plt.xlabel('Cluster Group')
plt.ylabel('Age')
sns.move_legend(plot1, 'upper left', bbox_to_anchor=(1,1))
plt.show()

#Creating box plot for Income and Cluster2 and ReAdmis
plot2 = sns.boxplot(data=df, x='Cluster2', y='Income', hue='ReAdmis')
plt.title('Readmission by Patient Income and Cluster Group')
plt.xlabel('Cluster Group')
plt.ylabel('Income')
sns.move_legend(plot2, 'upper left', bbox_to_anchor=(1,1))
plt.show()

#Creating bar plot for Age and Cluster 2 and ReAdmis
plot3 = sns.barplot(data=df, x='Cluster2', y='Age', hue='ReAdmis')
plt.title('Readmission by Age and Cluster Group')
plt.xlabel('Cluster Group')
plt.ylabel('Age')
sns.move_legend(plot3, 'upper left', bbox_to_anchor=(1,1))
plt.show()

#Creating bar plot for Income and Cluster 2 and ReAdmis
plot4 = sns.barplot(data=df, x='Cluster2', y='Income', hue='ReAdmis')
plt.title('Readmission by Income and Cluster Group')
plt.xlabel('Cluster Group')
plt.ylabel('Income')
sns.move_legend(plot4, 'upper left', bbox_to_anchor=(1,1))
plt.show()
```

#### E1. Accuracy of Clustering Technique<a id='e1'></a>

I created two different K-means clustering models for the analysis, so I will discuss each separately.

To measure the accuracy of the clustering, I used the silhouette score, which ranges from -1 to 1.  A score of -1 indicates that the clusters are not clearly distinguished, and a score of 1 means that the clusters are distinguished.  A score of 0 means that the difference between clusters is not significant (Bhardwaj, 2020).

<u>4-Cluster, 3-Feature Model</u></br>

This was the first model I created for this analysis, and it included 3 features (`Age`, `Income`, and `Doc_visits`) and used 4 clusters.  The silhouette score was 0.28, which indicates that the clustering has slight differences between clusters.

<u>3-Cluster, 2-Feature Model</u></br>

The second model I created for this analysis used 3 clusters and included 2 features (`Age` and `Income`).  The silhouette score was 0.42, which still indicates that the clustering has slight differences between clusters but an improvement over the previous model.

The second model created more clearly distinguishable clusters based on the silhouette scores.

#### E2. Results and Implications<a id='e2'></a>

The k-means clustering model's silhouette score of 0.42 indicates that the model can somewhat effectively cluster the data into groups of similar age and income.  Although the score is closer to 0 than 1, the second model had an improved score over the first model.  The analysis outcome is essential to the hospital's goal of decreasing the readmission frequency rate since they should be able to group incoming patients into different clusters and determine if they are more likely to be readmitted.

#### E3. Limitation<a id='e3'></a>

A limitation of the analysis is the imbalance in the available data.  The mean age across all 10,000 patients is 53.5, and the first quartile is 36; 75% are in their mid to late thirties or older.  The age range in the data set is 18 to 89 years old, possibly contributing to why the mean age is so high.  The model's effectiveness would be limited to patients within this age range.

Another area for improvement in the data analysis is that more information is needed regarding the 'ReAdmis' feature. The data definitions don't specify if 'Yes' indicates that readmission is related to ongoing issues from the initial admission. If, for example, the patient is readmitted for a non-related medical issue, but the hospital flags them as a readmitted patient, this analysis will misinterpret the results, and vice-versa, if the definition is that `ReAdmis` would be classified as `No` if the medical reason is related to the initial admission.
