### PART II: METHOD JUSTIFICATION

#### B1. Explanation of Classification Method<a id='b1'></a>

I initially leaned towards selecting the decision tree model but decided on the random forest method because it is a "more powerful descendant" of the decision tree (Bruce et al., 2020, p.249). Each tree makes its prediction, and then all of the predictions from each tree are averaged to produce a single prediction (Elleh, 2020, Slide 37). The averaging of all the decision trees makes the random forest a better classification method.

A decision tree is a non-parametric supervised machine learning method that separates the data into leaf and decision nodes through branches. Each tree is a collection of "if-else" questions in the form of a decision node that splits into more decision nodes and/or leaf nodes. For example, a decision node could be "Is the individual a male?" If the answer is `No`, then the leaf node would end at `Female`, but if the answer is `Yes`, the next layer could be another decision node, like "Does the individual have a beard?" There could be two leaf nodes for a `Yes`/`No` response, or the decision tree branch could lead to another decision node. Each tree makes its independent predictions to label a target variables. The prediction of each tree flows to the random forest model for a final prediction. For a regression model, the random forest averages the predictions of the best estimators; this usually leads to less variance than what's observed in individual tree predictions (Kawerk, n.d.).

I expect to see `TotalCharge` as a leading predictor through this method since, practically speaking, at least, a patient would incur a higher cost the longer they remain in medical care.

```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn import metrics
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score, make_scorer

# Importing medical_clean.csv file into Jupyter
df = pd.read_csv(r'Files\medical_clean.csv')

# Output display settings
pd.set_option('display.max_columns', None)

# Checking data using the .shape
print('Data Set Summary:')
print(f'The data set has {df.shape[0]} rows and {df.shape[1]} columns.')
print('-'*100)

# Checking data using the .info() function
print(df.info())
print('-'*100)

# Checking data for null values using the .isnull().sum() function
print('Null Value Review')
print('-'*23)
print(df.isnull().sum())
print('-'*100)

# Checking data for duplictes
print('Duplicates Review')
print('-'*17)
print(df.duplicated().value_counts())
```

```python
# Dropping features not needed for model
df = df.drop(columns=['CaseOrder', 'City', 'County', 'Customer_id', 'Interaction', 'Job', 
                      'Lat', 'Lng', 'State', 'TimeZone', 'UID', 'Zip'])

# Renamed 'Item' features
df.rename(columns={'Item1': 'TimelyAdmission'}, inplace=True)
df.rename(columns={'Item2': 'TimelyTreatment'}, inplace=True)
df.rename(columns={'Item3': 'TimelyVisit'}, inplace=True)
df.rename(columns={'Item4': 'Reliability'}, inplace=True)
df.rename(columns={'Item5': 'Options'}, inplace=True)
df.rename(columns={'Item6': 'HoursofTreatment'}, inplace=True)
df.rename(columns={'Item7': 'CourteousStaff'}, inplace=True)
df.rename(columns={'Item8': 'ActiveListening'}, inplace=True)

# Creating list of continuous variables
num_var = ['Additional_charges', 'Age', 'Children', 'Doc_visits', 'Full_meals_eaten', 'Income', 'Population', 
           'TotalCharge', 'VitD_levels', 'vitD_supp', 'TimelyAdmission', 'TimelyTreatment', 'TimelyVisit', 
           'Reliability', 'Options', 'HoursofTreatment', 'CourteousStaff', 'ActiveListening']
# Creating target variable
target_var = ['Initial_days']

# Encoding 'Complication_risk' to 0/1/2
df['Complication_risk'] = df['Complication_risk'].map({'Low': 0, 'Medium': 1, 'High': 2})

# Creating list of remaining categorical variables that need to be changed from Yes/No to 1/0
reex_cat_var = ['ReAdmis','Soft_drink', 'HighBlood', 'Stroke', 'Overweight', 'Arthritis', 'Diabetes', 
                     'Hyperlipidemia', 'BackPain', 'Anxiety', 'Allergic_rhinitis', 'Reflux_esophagitis', 'Asthma']

for item in reex_cat_var:
    df[item] = df[item].map({'No': 0, 'Yes': 1})
    
# Creating list of categorical variables that will need one hot encoding
hot_cat_var = ['Area', 'Marital', 'Gender', 'Initial_admin', 'Services']

# The code used to combine the lists is from the website Tutorials Point (Chiluka, 2022)
for item in hot_cat_var:
    results = pd.get_dummies(df[item], drop_first=True)
    for a in results:
        df[a] = results[a]
    del df[item]
    
# Creating list of categorical variables
cat_var = df.drop(columns=num_var).drop(columns=target_var).columns.tolist()

# Generating statistical summaries
print(df[num_var].describe().T)
print('-'*100)

# Checking for value counts for all categorical variables
for item in cat_var:
    print(f'{item} Count')
    print(df[item].value_counts())
    print('-'*100)
```

```python
# Creat heatmap to inspect correlation of quantitative before feature selection
corr = df[num_var].corr()
plt.figure(figsize=(12,10))
sns.heatmap(data=corr, annot=False, cmap='Spectral', vmin=-1, vmax=1)
plt.title('Heatmap for Quantitative Features')
plt.show()
print('-'*120)

# Creat heatmap to inspect correlation of qualitative before feature selection
plt.figure(figsize=(12,10))
sns.heatmap(data=df[cat_var].corr(), annot=False, cmap='Spectral', vmin=-1, vmax=1)
plt.title('Heatmap for Qualitative Features')
plt.show()
```

```python
# Dropping features that are closely correlated
df = df.drop(columns=['Additional_charges', 'TimelyAdmission', 'TimelyTreatment', 'TimelyVisit', 
                      'Reliability', 'Options', 'HoursofTreatment', 'CourteousStaff', 'ActiveListening',
                      'Married', 'Never Married', 'Separated', 'Widowed', 'Emergency Admission', 
                      'Observation Admission', 'CT Scan', 'Intravenous', 'MRI'])

# Creating variables to assign features and target variables to x and y, respectively (Elleh, 2020, Slide 28)
x = df.drop('Initial_days', axis=1, inplace=False)
y = df['Initial_days']

# Creating the feature selection variable, got code for 'f_regression' from scikitlearn.com
var_select = SelectKBest(score_func=f_regression, k='all')
var_select.fit_transform(x, y)

# Finding the p-values to select variables with value < 0.05
p_values = pd.DataFrame({'Feature': x.columns, 'p_value': var_select.pvalues_}).sort_values('p_value')
p_values[p_values['p_value'] < 0.1].sort_values('p_value', axis=0, ascending=False)
```

```python
# Check multicollinearity using VIF
b = df[['Full_meals_eaten', 'BackPain', 'Population', 'Arthritis', 'Children', 'ReAdmis', 'TotalCharge']]

vif_data = pd.DataFrame()
vif_data['Feature'] = b.columns
vif_data['VIF'] = [variance_inflation_factor(b.values, i) for i in range(len(b.columns))]
vif_data.sort_values('VIF', axis=0, ascending=False, inplace=True)
vif_data
```

### PART IV: ANALYSIS

#### D1. Splitting the Data<a id='d1'></a>

```python
# Split data into training and testing sets (Elleh, 2020, Slide 30)
x = df.drop('Initial_days', axis=1, inplace=False)
y = df['Initial_days']

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=15)

# Saving training and testing files to CSV
x_train.to_csv(r'Files/Task 2/x_train.csv', index=False)
x_test.to_csv(r'Files/Task 2/x_test.csv', index=False)
y_train.to_csv(r'Files/Task 2/y_train.csv', index=False)
y_test.to_csv(r'Files/Task 2/y_test.csv', index=False)
```

#### D2. Output and Intermediate Calculations<a id='d2'></a>

To perform the random forest analysis, I used the `RandomForestRegressor` from `sklearn` and defined `rf`.  I trained `rf` by fitting it to the training data set and then predicted the label on the test data set, `y_pred`.

Parameters are learned from the data, but hyperparameters of machine learning models are tuned to improve the model's performance and should be set before training the data.  (Kawerk, n.d.).  The goal of setting hyperparameters is to get the best possible score.  To set the hyperparameters, I defined the parameters (`params`) and the scoring model that should be used (`scorer`), then I passed these settings through `GridSearchCV` package from `sklearn`.

After instantiating the `grid`, I trained it by fitting it to the training data set and then extracted the best possible parameters.  I also predicted the label on the test data set using `grid` and created a new prediction called `y_train_pred`.

Next, I calculated the mean squared error, root mean squared error and r<sup>2</sup> for the training and testing data sets to compare the results.

The last step was to check the feature importance by plotting the features on a horizontal bar graph.

I used slides 40-43 from Dr. Elleh's <i>D209 Data Mining I Task 2 Cohort</i> PowerPoint presentation to guide me in creating the code in this part.

```python
# Hyperparameter tuning for random forest
params = {'n_estimators': [10,50,100],
              'max_features': [2,3,4],
              'max_depth': [8, None]
             }

# Instantiate the RandomForestRegressor object
rf = RandomForestRegressor(random_state=15)

# Fit rf to training set
rf.fit(x_train, y_train)

# Predict test set labels 'y_pred'
y_pred = rf.predict(x_test)

# Defining 'scorer'
scorer = make_scorer(MSE)

# Instantiate the grid search model
grid = GridSearchCV(estimator=rf,
                    param_grid=params,
                    cv=10, scoring=scorer,
                    n_jobs=-1)

# Fit grid to training set
grid.fit(x_train, y_train)

# Extract best parameters from 'grid'
print('Best params: ',grid.best_params_)
print('-'*100)

# Defining 'y_train_pred'
y_train_pred = grid.predict(x_train)

# Checking scores for top performing model
print('Training Score (MSE): {:.3f}'.format(grid.best_score_))
print('Training Score (RMSE): {:.3f}'.format(grid.best_score_**(1/2)))
print('Training Score (R-Squared): {:.3f}'.format(r2_score(y_train, y_train_pred)))

# Checking prediction accuracy
y_pred = grid.predict(x_test)
print('-'*100)
print('Testing Score (MSE): {:.3f}'.format(MSE(y_test, y_pred)))
print('Testing Score (RMSE): {:.3f}'.format(MSE(y_test, y_pred)**(1/2)))
print('Testing Score (R-Squared): {:.3f}'.format(r2_score(y_test, y_pred)))
```

```python
# Create a pd.Series for feature importance
importance = pd.Series(data=grid.best_estimator_.feature_importances_, index=x_train.columns)

# Sorting 'importance'
sorted_importance = importance.sort_values()

# Plotting feature sorted_importance
sorted_importance.plot(kind='barh', color='lightgreen')
plt.title('Features Importance')
plt.show()
```
